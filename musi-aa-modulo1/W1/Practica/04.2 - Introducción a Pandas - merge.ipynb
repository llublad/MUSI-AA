## Cargar datos y primera inspección

```python
import pandas as pd
```

```python
df_iris = pd.read_csv('data/iris_dataset.csv')
df_iris_recolectors = pd.read_csv('data/iris_recolectors.csv',
                                  dtype={'rec_id': str})
df_recolectors_hours= pd.read_csv('data/recolectors_hours.csv')
```

```python
## Qué número único de registry_id tenemos en cada dataframe ?

print("Número de registry_id, únics a df_iris:",
      len(df_iris['registry_id'].unique()))

print("Número de registry_id, únics a df_iris_recolectors:",
      len(df_iris_recolectors['registry_id'].unique()))

print("Número de id, únics a df_recolectors_hours:",
      len(df_recolectors_hours['id'].unique()))
```

Crea un dataframe nuevo cruzando los dos dataframes por la columna `registry_id`

```python
df_iris_w_info_rec = df_iris.merge(
    right=df_iris_recolectors, 
    on = 'registry_id') ## your code goes here
```

```python
df_iris_w_info_rec
```

Si no se especifica, el tipo de cruze es "inner", la intersección entre los dos dataframes

Repite la operación haciendo un left join de df_iris sobre df_iris_recolectors

```python
df_iris_w_info_rec = df_iris.merge(
    right=df_iris_recolectors, 
    on = 'registry_id',
    how = 'left') ## your code goes here
```

```python
df_iris_w_info_rec
```

Usa la funcionalidad groupby para contar el numero de clases de iris por cada recolector de datos (rec_id)

```python
df_num_datapoints_per_rec = df_iris_w_info_rec.groupby(['rec_id'])
df_num_datapoints_per_rec[['registry_id']].count()
## your code goes here
```

```python
df_num_datapoints_per_rec
```

```python
df_recolectors_hours
```

Convierte la columna id a tipo string

```python
df_recolectors_hours['id'] = df_recolectors_hours['id'].apply(str) ## your code goes here
```

```python
df_recolectors_hours['id'].dtype
```

Haz un merge entre el df_num_datapoints_per_rec y el df_recolector_hours por el id

```python
df_num_datapoints_hours = df_num_datapoints_per_rec[['registry_id']].count().merge(
    right = df_recolectors_hours, 
    left_on = 'rec_id',
    right_on = 'id',
    how = 'left'
    ) ## your code goes here
df_num_datapoints_hours
```

Calcula el tiempo medio de recolección por datapoint

```python
df_num_datapoints_hours['mean_recolection_time'] = \
    df_num_datapoints_hours['hours'] / df_num_datapoints_hours['registry_id'] ## your code goes here
```
